```python
# 必要なライブラリをインポート
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

# Step1: Excelファイルの読み込み
df_2022 = pd.read_excel('2022_年間売上表.xlsx', sheet_name='Sheet1')
df_2023 = pd.read_excel('2023_年間売上表.xlsx', sheet_name='Sheet1')

# Step2: データの連結
df = pd.concat([df_2022, df_2023], ignore_index=True)

# Step3: データの集計 - 「商品」「売上年」ごとにグループ化し、金額の合計を算出
df_grouped = df.groupby(['商品', '売上年'])['金額（千円）'].sum().reset_index()

# Step4: 集計結果を新規Excelファイル「売上集計表.xlsx」に書き込む
df_grouped.to_excel('売上集計表.xlsx', index=False, sheet_name='集計結果')

# Step5: openpyxlを使ってヘッダー部分の書式を変更
wb = load_workbook('売上集計表.xlsx')
ws = wb['集計結果']

# ヘッダー行（1行目）を薄いグレーに設定
fill = PatternFill(start_color="F2F2F2", end_color="F2F2F2", fill_type="solid")

for cell in ws[1]:
    cell.fill = fill

# 保存して終了
wb.save('売上集計表.xlsx')

# 最後にデータフレームを表示
df_grouped
```
