```python
# 必要なライブラリをインポート
import pandas as pd

# 売上データの作成
data = {
    '日付': ['2023-05-17', '2023-05-18', '2023-05-19', '2023-05-20', '2023-05-21'],
    '社員名': ['山田', '佐藤', '鈴木', '田中', '高橋'],
    '売上': [100, 200, 150, 300, 250],
    '部門': ['メーカー', '代理店', 'メーカー', '商社', '代理店']
}

# DataFrameを作成
df = pd.DataFrame(data)

# 全体の平均売上を計算し、小数点以下を切り捨てて整数に変換
average_sales = int(df['売上'].mean())

# 平均売上列を追加
df['平均売上'] = average_sales

# 業績ランクを決定する関数を定義
def performance(row):
    if row['売上'] >= average_sales + 50:
        return 'A'
    elif row['売上'] >= average_sales:
        return 'B'
    else:
        return 'C'

# 「業績ランク」列を追加
df['業績ランク'] = df.apply(performance, axis=1)

# 結果を新規ファイル「業績.xlsx」に保存
df.to_excel('業績.xlsx', index=False)

# 結果を表示
df
```
